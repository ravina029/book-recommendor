```python
ratings=pd.read_csv('Ratings.csv')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={"User-ID":"user_id",
                        "Book-Rating":"rating"},inplace=True)
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].unique().shape
```

```python
a=ratings['user_id'].value_counts()>100
```

```python
a[a].shape
```

```python
b=a[a].index
```

```python
ratings=ratings[ratings['user_id'].isin(b)]
```

```python
ratings.shape
```

```python
books_with_ratings=ratings.merge(df_books, on='ISBN')
```

```python
books_with_ratings.head()
```

```python
rating_counts=books_with_ratings.groupby('Book')['rating'].count().reset_index()
```

```python
rating_counts.head()
```

```python
final_ratings=books_with_ratings.merge(rating_counts, on='Book')
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_rating=final_ratings[final_ratings['rating_y']>50]
```

```python
final_rating.shape
```

```python
final_rating.sample(20)
```

```python
print(final_rating.duplicated('user_id').sum())
print(final_rating.duplicated('Book').sum())
```

```python
final_rating.drop_duplicates(['user_id','Book'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='Book',values='rating_x')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```
